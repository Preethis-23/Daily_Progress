# Leet Code 2027
Minimum moves to convert to string

```python
class Solution:
    def minimumMoves(self, s: str) -> int:
        cnt=0
        i=0

        while i<len(s):
            if s[i]=="X":
                cnt+=1
                i+=3
            else:
                i+=1
        return cnt
```

# Leet Code 300
Longest Increasing Subsequence

```python
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        dp=[1 for _ in range(len(nums))]

        for i in range(len(nums)):
            for j in range(i):
                if nums[j]<nums[i]:
                    dp[i]=max(dp[i],dp[j]+1)
        return max(dp)

#another method ( 2 for loop ), like mark index of array elements and comapare index to gain count (i.e,) [4,1,10,5] like [1,0,3,2] check index to collect LIS
        
```

# Leet Code 169
Majority Element

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        
        # O(n log n)
        # O(log n)
        if len(nums)==1:
            return nums[0]
        mid=len(nums)//2

        left_part=self.majorityElement(nums[:mid])
        right_part=self.majorityElement(nums[mid:])

        if left_part==right_part:
            return left_part
        
        left_cnt=nums.count(left_part)
        right_cnt=nums.count(right_part)

        result=0
        if left_cnt>right_cnt:
            result=left_part
        else:
            result = right_part
        return result if nums.count(result)>len(nums)//2 else 0



        #boyer moore voting algorithm
        # time O(n) space O(1)
        arr=nums
        candidate = None
        count = 0

        # Find candid
        for num in arr:
            if count == 0:
                candidate = num
            count += (1 if num == candidate else -1)

        #  Verify candid
        if arr.count(candidate) > len(arr) // 2:
            return candidate
        else:
            return "NO"
        
```

Partitioning Into Minimum Number Of Deci-Binary Numbers

# Leet Code 200
Number of Islands

```python
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:

        row,col = len(grid), len(grid[0])
        cnt=0

        def dfs(r,c):
            if r<0 or c<0 or r>=row or c>=col:
                return
            if grid[r][c]=='0':
                return
            grid[r][c]='0'

            dfs(r+1,c)
            dfs(r-1,c)
            dfs(r,c+1)
            dfs(r,c-1)
        for i in range(row):
            for j in range(col):
                if grid[i][j]=='1':
                    cnt+=1
                    dfs(i,j)
        return cnt
    
#its just like take a start of land and count +1 then spread as long as possible then find next start of land and so on.....
```

# Leet Code 621
Task Scheduler

```python
class Solution:
    def leastInterval(self, tasks: List[str], n: int) -> int:
        
        freq = Counter(tasks)
        
        maxFreq = max(freq.values())
        countMax = list(freq.values()).count(maxFreq)
        
        return max(len(tasks), (maxFreq - 1) * (n + 1) + countMax)
```
