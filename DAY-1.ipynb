# Leet Code 78
Subsets

```python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        result=[]
        path=[]

        def backtrack(start):
            result.append(path.copy())

            for i in range(start,len(nums)):
                path.append(nums[i])
                backtrack(i+1)
                path.pop()
        backtrack(0)
        return result

        
```

Subsets II

```python
class Solution:
    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        path=[]
        result=[]
        #checker=[]

        def backtrack(start):
            #if sorted(path) not in checker:
            result.append(path.copy())
             #   checker.append(sorted(path))

            for i in range(start,len(nums)):
                if i > start and nums[i] == nums[i-1]:
                    continue
                path.append(nums[i])
                backtrack(i+1)
                path.pop()
        backtrack(0)
        return result        
```

# Leet code 39
Combination Sum I

```python

class Solution:

    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        nums=candidates
        result=[]
        path=[]
        idx=0
        total=0

        def backtrack(start):
            nonlocal total
            if total==target:
                result.append(path.copy())
                return
            if total>target:
                return False

            for i in range(start,len(nums)):
                path.append(nums[i])
                total+=nums[i]
                backtrack(i)
                path.pop()
                total-=nums[i]
            return result
        backtrack(0)
        return result

            
        
```

# Leet Code 40
Combination Sum II

```python
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        nums=candidates
        nums.sort()
        result=[]
        path=[]
        total=0

        def backtrack(start):
            nonlocal total
            if total==target:

                result.append(path.copy())
                return
            if total>target:
                return False
            for i in range(start, len(nums)): # we use to start to prevent the formation 2,1,5 from 1,2,5( after creating 1,2,5 recursio forgots there exists value 1 , so there will be no 2,15)
                if i>start and nums[i]==nums[i-1]: #eventhough current idx>start idx:if current and prev idx's values are same skip it ( array is sorted already)
                    continue
                total+=nums[i]
                path.append(nums[i])

                backtrack(i+1)

                path.pop()
                total-=nums[i]
            return 
        backtrack(0)
        return result
            
                
        
```

# Leet Code 216 
Combination Sum III

```python
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        nums=[1,2,3,4,5,6,7,8,9]
        result=[]
        path=[]
        total=0

        def backtrack(start):
            nonlocal total
            if len(path)==k and total==n: #goal checking
                result.append(path.copy())
                return
            if total>n or len(path)>k: #pruning
                return False
            for i in range(start,len(nums)):
                if nums[i] not in result:
                    path.append(nums[i])
                    total+=nums[i]

                    backtrack(i+1)

                    path.pop()
                    total-=nums[i]
            return
        backtrack(0)
        return result
```
