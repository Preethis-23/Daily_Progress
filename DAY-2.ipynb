# Leet Code 1137
N-th Tribonacci Number

```python
class Solution:
    def tribonacci(self, n: int) -> int:

        if n==0:
            return 0
        if n==1 or n==2:
            return 1

        #define state
        dp = [0]*(n+1)

        #base case
        dp[0] = 0
        dp[1] = 1
        dp[2] = 1
        
        #recurrence relation
        for i in range(3, n+1):
            dp[i] = dp[i-1] + dp[i-2] + dp[i-3]

        #return output
        return dp[n]

        
        
```

# Leet Code 746
Min Cost Climbing Stairs

```python
class Solution:
    def minCostClimbingStairs(self, cost: List[int]) -> int:
        
        n=len(cost)
        #define state
        dp = [float('inf')]*(n)

        #base case
        dp[0]=cost[0]
        dp[1]=cost[1]

        #recurrence relation
        for i in range(2,n):
            dp[i] = cost[i] + min(dp[i-1], dp[i-2])

        #return output
        return min(dp[n-1], dp[n-2]) #here in this problem min cost is calculated for two paths, n-1 and n-2, so we have to find minimum
```

# Leet Code 53
Maximum Subarray

```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        n=len(nums)

        #define state
        dp = [0]*n

        #base case
        dp[0] = nums[0]

        #recurrence relation
        for i in range(1,n):
            dp[i]=max(nums[i],dp[i-1]+nums[i])

        #return output
        return max(dp)
```

# Leet Code 152
