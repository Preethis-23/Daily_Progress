# QSN 1
The maze is given using a binary matrix, where it is marked with 0, it is a valid path, otherwise 1 for a blocked cell.

NOTE: The rat can only move in two directions, either to the right or to the down.

```python
n=int(input())
lst = [list(map(int, input().split())) for _ in range(n)]
visited = [[0]*n for _ in range(n)]
moves = [(0,1), (1,0)]
start = (0,0)
target = (n-1, n-1)

def backtrack(r, c):
    if r<0 or c<0 or r>=n or c>=n:
        return False
    if visited[r][c]==1 or lst[r][c]==1:
        return False
    if (r, c)==target:
        visited[r][c] = 1
        return True
    
    visited[r][c] = 1
    
    for mr, mc in moves:
        nr, nc = r+mr, c+mc
        if backtrack(nr, nc):
            return True
    visited[r][c] = 0
    return False
    
if backtrack(start[0], start[1]):
    for row in visited:
        print(*row)
else:
    print("No solution exists")
            
```

# QSN 2
Minimum steps to reach destination

```python
n=int(input())
lst = [list(map(int, input().split())) for _ in range(n)]
visited = [[0]*n for _ in range(n)]
moves = [(0,1), (1,0), (-1, 0), (0, -1)]
start = (0,0)
target = (n-1, n-1)
short = float("inf")
def backtrack(r, c, steps):
    global short
    #goal
    if (r, c)==target:
        steps+=1
        short = min(short, steps)
        visited[r][c] = 1
        return True

    #valid
    if r<0 or c<0 or r>=n or c>=n:
        return False
    if visited[r][c]==1 or lst[r][c]==1:
        return False
    
    
    visited[r][c] = 1
    
    for mr, mc in moves:
        nr, nc = r+mr, c+mc
        if backtrack(nr, nc, steps+1):
            return True
    visited[r][c] = 0
    return False
    
if backtrack(start[0], start[1],0):
    print("Minimum Steps:", short)
    for row in visited:
        print(*row)
```
