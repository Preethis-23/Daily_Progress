# Leet Code 64
Minimum Path Sum

```python
class Solution:
    def minPathSum(self, grid: List[List[int]]) -> int:
        m=len(grid)
        n=len(grid[0])
        #define state
        dp = [[0]*(n) for _ in range(m)]

        #base case
        dp[0][0]=grid[0][0]

        #recurrence relation
        for i in range(1,m):
            dp[i][0] = dp[i-1][0] + grid[i][0]
        for j in range(1,n):
            dp[0][j] = dp[0][j-1] + grid[0][j]

        for i in range(1, m):
            for j in range(1, n):
                dp[i][j]=grid[i][j] + min(dp[i][j-1], dp[i-1][j])
        return dp[m-1][n-1]
        
```
