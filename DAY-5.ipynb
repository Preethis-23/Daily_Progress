# Leet Code 322
Coin Change

```python
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        dp = [float("inf")] * (amount + 1)
        dp[0] = 0

        for i in range(1, amount + 1):
            for coin in coins:
                if coin <= i:
                    dp[i] = min(dp[i], dp[i - coin] + 1)

        return dp[amount] if dp[amount] != float("inf") else -1


'''
consider given amount=7, coins = [1, 3, 5]

for i range 1 to 7
for coin in coins ( no of coin present is no of possibility to form amt i, make sure only if coin<=i)

assume, 7's possiblilites
7-1=6=>dp[6]+1 #here +1 denotes current coin used once, and need 6 money 
7-3=4=>dp[4]+1 #4 money needed
7-5=2=>dp[2]+1 #2 money needed

for dp[6, 4, 2] we would have already found the value like
6=> 6-1=dp[5]+1, 6-3=dp[3]+1, 6-5=dp[1]+1
4=> 4-1=dp[3]+1, 4-2=dp[2]+1
2=> 2-1=dp[1]+1
'''
```

# Leet Code 416
Partition Equal Subset Sum

```python
class Solution:
    def canPartition(self, nums: List[int]) -> bool:

        n=sum(i for i in nums)
        if n%2!=0:
            return False
        else:
            n=n//2
            #define state
            dp = [False]*(n+1)

            #base case
            dp[0] = True

            #recurrence relation
            for i in range(len(nums)):
                for j in range(n,nums[i]-1,-1):
                    if dp[j-nums[i]]==True:
                        dp[j] = True
            return dp[n]
        
```

# Leet Code 412
FizzBuzz


```python
class Solution:
    def fizzBuzz(self, n: int) -> List[str]:
        arr = [""]*n
        for i in range(1,n+1):
            if i%3==0 and i%5==0:
                arr[i-1] = "FizzBuzz"
            elif i%3==0:
                arr[i-1] = "Fizz"
            elif i%5 == 0:
                arr[i-1] = "Buzz"
            else:
                arr[i-1] = str(i)
        return arr
```
