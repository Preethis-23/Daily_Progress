# Leet Code 1029
Two City Scheduling

```python
class Solution:
    def twoCitySchedCost(self, costs: List[List[int]]) -> int:
        costs.sort(key=lambda x:x[0]-x[1])

        n=len(costs)//2
        tot = 0

        for i in range(len(costs)):
            if i<n:
                tot += costs[i][0]
            else:
                tot += costs[i][1]
        return tot
        
```

# Leet Code 118
Pascal's triangle

```python
class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
        
        dp = [[0]*i for i in range(1,numRows+1)]

        dp[0][0] = 1


        for i in range(1,numRows):
            for j in range(i+1):
                if j==0:
                    dp[i][j] = dp[i-1][j]
                elif j==i:
                    dp[i][j] = dp[i-1][j-1]
                else:
                    dp[i][j] = dp[i-1][j-1] + dp[i-1][j]
        return dp
        
```
